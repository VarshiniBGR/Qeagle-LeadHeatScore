## üìà Performance Metrics Overview


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# Load our perfect metrics
metrics_df = pd.read_csv('../metrics/results_metrics.csv')
print("üìä LEAD HEATSCORE PERFORMANCE METRICS")
print("=" * 50)
print(metrics_df.to_string(index=False))

# Extract key metrics
f1_macro = metrics_df[metrics_df['metric'] == 'f1_macro']['value'].iloc[0]
accuracy = metrics_df[metrics_df['metric'] == 'accuracy']['value'].iloc[0]
latency = metrics_df[metrics_df['metric'] == 'latency_ms']['value'].iloc[0]

print(f"\nüèÜ KEY PERFORMANCE INDICATORS:")
print(f"F1 Score (Macro): {f1_macro:.3f}")
print(f"Accuracy: {accuracy:.1f}%")
print(f"Average Latency: {latency:.1f}ms")
print(f"Status: {'‚úÖ PASS' if f1_macro >= 0.80 else '‚ùå FAIL'}")
```

## üîÑ Model Improvement Journey

### Before vs After Comparison


```python
# Model improvement comparison
comparison_data = {
    'Model': ['Logistic Regression', 'Random Forest'],
    'F1 Score': [0.336, 1.000],
    'Accuracy (%)': [49.0, 100.0],
    'Status': ['FAIL', 'PASS'],
    'Improvement': ['Baseline', '+198%']
}

comparison_df = pd.DataFrame(comparison_data)
print("üîÑ MODEL IMPROVEMENT COMPARISON")
print("=" * 40)
print(comparison_df.to_string(index=False))

# Visualize the improvement
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# F1 Score comparison
models = ['Logistic\nRegression', 'Random\nForest']
f1_scores = [0.336, 1.000]
colors = ['#ff6b6b', '#51cf66']

bars1 = ax1.bar(models, f1_scores, color=colors, alpha=0.8, edgecolor='black', linewidth=1)
ax1.axhline(y=0.80, color='red', linestyle='--', linewidth=2, label='Target (0.80)')
ax1.set_ylabel('F1 Score', fontsize=12, fontweight='bold')
ax1.set_title('F1 Score Improvement', fontsize=14, fontweight='bold')
ax1.set_ylim(0, 1.1)
ax1.legend()

# Add value labels on bars
for bar, score in zip(bars1, f1_scores):
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height + 0.02,
             f'{score:.3f}', ha='center', va='bottom', fontweight='bold')

# Accuracy comparison
accuracy_scores = [49.0, 100.0]
bars2 = ax2.bar(models, accuracy_scores, color=colors, alpha=0.8, edgecolor='black', linewidth=1)
ax2.axhline(y=85, color='red', linestyle='--', linewidth=2, label='Target (85%)')
ax2.set_ylabel('Accuracy (%)', fontsize=12, fontweight='bold')
ax2.set_title('Accuracy Improvement', fontsize=14, fontweight='bold')
ax2.set_ylim(0, 110)
ax2.legend()

# Add value labels on bars
for bar, score in zip(bars2, accuracy_scores):
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height + 2,
             f'{score:.1f}%', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

print(f"\nüéâ DRAMATIC IMPROVEMENT ACHIEVED!")
print(f"F1 Score: +{((1.000 - 0.336) / 0.336 * 100):.0f}% improvement")
print(f"Accuracy: +{((100.0 - 49.0) / 49.0 * 100):.0f}% improvement")
```

```python
# Test RAG System Performance
import requests
import json

def test_rag_system():
    """Test the RAG system performance."""
    try:
        # Test debug endpoint
        response = requests.get("http://localhost:8000/api/v1/admin/debug-rag")
        if response.status_code == 200:
            data = response.json()
            
            print("ü§ñ RAG SYSTEM PERFORMANCE TEST")
            print("=" * 40)
            
            # Circuit breaker status
            circuit_breakers = data.get('circuit_breakers', {})
            print(f"Circuit Breakers: {circuit_breakers}")
            
            # RAG test results
            rag_test = data.get('rag_test', {})
            print(f"Query: {rag_test.get('query', 'N/A')}")
            print(f"Results Count: {rag_test.get('results_count', 0)}")
            
            # Embedding test
            embedding_test = rag_test.get('embedding_test', {})
            print(f"Embedding Success: {embedding_test.get('success', False)}")
            print(f"Embedding Dimensions: {embedding_test.get('dimensions', 0)}")
            
            # Status
            status = data.get('status', 'unknown')
            print(f"Status: {status}")
            
            return data
        else:
            print(f"‚ùå Error: {response.status_code}")
            return None
            
    except Exception as e:
        print(f"‚ùå Error testing RAG system: {e}")
        return None

# Test the RAG system
rag_results = test_rag_system()
```
